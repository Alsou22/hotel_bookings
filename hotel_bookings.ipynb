Проанализируйте даты запланированного прибытия. – На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?– Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

```python
df4 = df \
    .query("is_canceled == 0 and arrival_date_year == 2016 ") \
    .groupby(['arrival_date_month'], as_index = False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending = False)
df4.head(12)
# в 2016 г. чаще всего успешно оформляли бронь в октябре
```

```python
df5 = df \
    .query("is_canceled == 0 and arrival_date_year == 2017 ") \
    .groupby(['arrival_date_month'], as_index = False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending = False)
df5.head(12)
# 2017 г. наиболее успешный месяц - май
```

Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
df['total_kids'] = df.children + df.babies
df[['total_kids', 'children', 'babies']]
```

```python
df7 = df \
    .groupby('hotel', as_index = False) \
    .aggregate({'total_kids': 'mean'}) \
    .sort_values('total_kids', ascending = False)
df7
#среднее значение переменной оказалось наибольшим для отелей типа Resort Hotel
```

Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
df['has_kids'] = df['total_kids'] >=1
df['has_kids']
```

```python
df['has_kids'].value_counts()
```

```python
# Клиенты с детьми, отменившие бронирование
df8 = df \
    .query("has_kids == True and is_canceled == 1") \
    .groupby('has_kids', as_index = False) \
    .aggregate({'has_kids': 'count'}) 
df8
```

```python
# Процент оттока клиентов с детьми.Отношение клиентов с детьми отменивших бронирование, к общему количеству клиентов с детьми
churn_rate_t = df8/9332*100
churn_rate_t.round(2)
```

```python
# Клиенты без детей, отменившие бронирование
df9 = df \
    .query("has_kids == False and is_canceled == 1") \
    .groupby('has_kids', as_index = False) \
    .aggregate({'has_kids': 'count'}) 
df9
```

```python
# Процент оттока клиентов без детей.Отношение клиентов без детей отменивших бронирование, к общему количеству клиентов без детей
churn_rate_f = df9/110058*100
churn_rate_f.round(2)
```

Среди клиентов без детей показатель оттока выше
